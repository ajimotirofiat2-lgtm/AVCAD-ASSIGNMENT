### Generating a Random Visualization

Let's create some random data and visualize it using a scatter plot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data
np.random.seed(42) # for reproducibility
data = {
    'X': np.random.rand(50) * 100,
    'Y': np.random.rand(50) * 100,
    'Category': np.random.choice(['A', 'B', 'C'], 50)
}
df = pd.DataFrame(data)

print("Generated DataFrame head:")
display(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='X', y='Y', hue='Category', data=df, s=100, alpha=0.7)
plt.title('Random Scatter Plot of X vs Y by Category')
plt.xlabel('Random Value X')
plt.ylabel('Random Value Y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
