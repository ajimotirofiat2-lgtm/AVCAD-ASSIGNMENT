```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean plot style
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 100 # Set default figure DPI
```

### Loading the data and identifying top catchments

First, I'll load the `EFIplus_medit.csv` dataset and inspect its structure. Then, I'll determine the four catchments with the highest number of unique fish sampling sites.

```python
df = pd.read_csv('/content/EFIplus_medit.csv', sep=';')

# Check if the dataframe was incorrectly parsed into a single column
if len(df.columns) == 1 and ';' in df.columns[0]:
    print("\nWarning: The CSV was read as a single column. Attempting to re-parse.")
    full_header_string = df.columns[0]
    column_names = full_header_string.split(';')

    # Create a new DataFrame by applying the split operation to each row
    df = df[full_header_string].str.split(';', expand=True)
    df.columns = column_names

# Convert necessary columns to appropriate types after parsing
# For the operations below, 'Site_code', 'Catchment_name', and 'temp_ann' are critical.
# 'Site_code' should remain as string/object for unique counting.
# 'temp_ann' is expected to be numeric.
# 'Catchment_name' is expected to be a string.

# 'Site_code' should not be converted to numeric if it's a string identifier.
# 'Site_code' already appears to be an object type from initial loading if not re-parsed.
# If it does get converted during re-parsing, ensure it's not made numeric.
# For now, let's assume it's correctly loaded as object if the parsing works.

if 'temp_ann' in df.columns:
    df['temp_ann'] = pd.to_numeric(df['temp_ann'], errors='coerce')
else:
    print("Warning: 'temp_ann' column not found after parsing.")

print("Dataset Head:")
display(df.head())

print("\nDataset Info:")
df.info()

# Identify catchments with the highest number of fish sampling sites
# Assuming 'Site_code' refers to fish sampling sites.
if 'Catchment_name' in df.columns and 'Site_code' in df.columns:
    catchment_site_counts = df.groupby('Catchment_name')['Site_code'].nunique().sort_values(ascending=False)

    top_4_catchments = catchment_site_counts.head(4).index.tolist()

    print(f"\nTop 4 catchments by number of unique fish sampling sites: {top_4_catchments}")
    print("\nSite counts per catchment (top 10):")
    display(catchment_site_counts.head(10))
else:
    print("Error: 'Catchment_name' or 'Site_code' column not found after parsing, cannot identify top catchments.")
```

### Univariate Analysis of Annual Mean Temperature (`temp_ann`) for Top Catchments

Now, I'll create strip plots, histograms, and box plots for the `temp_ann` variable for each of the identified top four catchments. Each plot type will be presented in a single figure with 4 subplots to compare the distributions across catchments.

```python
# Filter data for the top 4 catchments
df_filtered = df[df['Catchment_name'].isin(top_4_catchments)].copy()

# Ensure the order of catchments for consistent plotting
df_filtered['Catchment_name'] = pd.Categorical(df_filtered['Catchment_name'], categories=top_4_catchments, ordered=True)

# --- Strip Plots ---
plt.figure(figsize=(15, 8))
plt.suptitle('Strip Plots of Annual Mean Temperature (temp_ann) by Catchment', fontsize=16)

for i, catchment_name in enumerate(top_4_catchments):
    plt.subplot(2, 2, i + 1)
    sns.stripplot(x=df_filtered[df_filtered['Catchment_name'] == catchment_name]['temp_ann'], jitter=0.2, alpha=0.7, color='skyblue')
    plt.title(f'Catchment: {catchment_name}')
    plt.xlabel('Annual Mean Temperature (temp_ann)')
    plt.ylabel('') # Remove y-label for cleaner look
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent suptitle overlap
plt.show()

# --- Histograms ---
plt.figure(figsize=(15, 8))
plt.suptitle('Histograms of Annual Mean Temperature (temp_ann) by Catchment', fontsize=16)

for i, catchment_name in enumerate(top_4_catchments):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df_filtered[df_filtered['Catchment_name'] == catchment_name]['temp_ann'], kde=True, bins=15, color='lightcoral')
    plt.title(f'Catchment: {catchment_name}')
    plt.xlabel('Annual Mean Temperature (temp_ann)')
    plt.ylabel('Frequency')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Box Plots ---
plt.figure(figsize=(15, 8))
plt.suptitle('Box Plots of Annual Mean Temperature (temp_ann) by Catchment', fontsize=16)

for i, catchment_name in enumerate(top_4_catchments):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x=df_filtered[df_filtered['Catchment_name'] == catchment_name]['temp_ann'], color='lightgreen')
    plt.title(f'Catchment: {catchment_name}')
    plt.xlabel('Annual Mean Temperature (temp_ann)')
    plt.ylabel('')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Additional Plot: Violin Plots

Violin plots combine aspects of box plots and kernel density plots, showing the distribution shape as well as summary statistics.

```python
plt.figure(figsize=(15, 8))
plt.suptitle('Violin Plots of Annual Mean Temperature (temp_ann) by Catchment', fontsize=16)

for i, catchment_name in enumerate(top_4_catchments):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x=df_filtered[df_filtered['Catchment_name'] == catchment_name]['temp_ann'], color='orchid')
    plt.title(f'Catchment: {catchment_name}')
    plt.xlabel('Annual Mean Temperature (temp_ann)')
    plt.ylabel('')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Evaluating the pros and cons of each plot type for univariate visualization

Here's a brief evaluation of the plots generated:

*   **Strip Plots**:
    *   **Pros**: Shows every data point, which is great for small datasets to see individual values and their density. Reveals multimodal distributions easily. No assumptions about the underlying distribution.
    *   **Cons**: Can become cluttered and hard to interpret with very large datasets. Overlapping points can obscure density unless jitter is applied.

*   **Histograms**:
    *   **Pros**: Clearly shows the shape of the distribution, including skewness, modality, and spread. Good for identifying outliers if bins are chosen appropriately.
    *   **Cons**: The appearance can be sensitive to the choice of bin size; too few bins can hide details, too many can show too much noise. Does not show individual data points.

*   **Box Plots**:
    *   **Pros**: Excellent for summarizing the five-number summary (minimum, first quartile, median, third quartile, maximum) and identifying potential outliers. Good for comparing distributions across multiple groups.
    *   **Cons**: Hides the actual shape of the distribution (e.g., cannot tell if it's bimodal). Does not show individual data points, only summaries.

*   **Violin Plots (Additional Plot)**:
    *   **Pros**: Combines the benefits of box plots (showing summary statistics like median and quartiles) with the density estimation of histograms/KDEs. Effectively shows the shape and spread of the distribution while also indicating central tendencies. Good for comparing distributions across groups.
    *   **Cons**: Can be more complex to interpret than a simple box plot for those unfamiliar with density plots. Like histograms, the kernel density estimation can be sensitive to bandwidth choice, though often less problematic than binning for histograms.

### Challenge: Mean `temp_ann` variation with sample size

Now, let's explore how the mean value of `temp_ann` varies with increasing sample sizes. I will take 1000 random samples with replacement for different sample sizes (10, 50, 100, 150, 200, 250, 300, 500, 1000), compute the mean `temp_ann` for each sample, and visualize the distribution of these sample means.

```python
if 'temp_ann' not in df.columns:
    print(f"Error: 'temp_ann' column not found in DataFrame. Available columns: {df.columns.tolist()}")
    raise KeyError("'temp_ann' column is missing after all prior processing.")

population_data = df['temp_ann'].dropna() # Using all available temp_ann data as population
population_mean = population_data.mean()
print(f"Population mean of temp_ann: {population_mean:.2f}")

sample_sizes = [10, 50, 100, 150, 200, 250, 300, 500, 1000]
num_samples = 1000

sample_means = []
sample_size_labels = []

for size in sample_sizes:
    for _ in range(num_samples):
        sample = population_data.sample(n=size, replace=True)
        sample_means.append(sample.mean())
        sample_size_labels.append(str(size))

sample_results_df = pd.DataFrame({
    'sample_mean': sample_means,
    'sample_size': sample_size_labels
})

plt.figure(figsize=(14, 7))
sns.boxplot(x='sample_size', y='sample_mean', data=sample_results_df, palette='viridis')
plt.axhline(y=population_mean, color='r', linestyle='--', label=f'Population Mean: {population_mean:.2f}')
plt.title('Distribution of Sample Means of temp_ann for Various Sample Sizes (1000 samples each)', fontsize=16)
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean of Annual Mean Temperature')
plt.legend()
plt.tight_layout()
plt.show()
```

### Interpretation of Sampling Plot

This plot visually demonstrates the Central Limit Theorem. As the sample size increases:

1.  **The distribution of sample means becomes more concentrated around the population mean.** The boxes (representing the interquartile range) become narrower, and the whiskers shorten.
2.  **The variability of sample means decreases.** This is evident from the shrinking spread of the box plots for larger sample sizes. The standard deviation of the sample means (standard error) decreases as $\frac{\sigma}{\sqrt{n}}$, where $\sigma$ is the population standard deviation and $n$ is the sample size.
3.  **The sample means converge towards the true population mean.** For larger sample sizes, the median of the sample means (the line inside the box) gets closer to the red dashed line representing the population mean.

From this visualization, we can observe that even with a sample size of 10, the distribution of sample means is centered around the population mean, but with high variability. As the sample size reaches around 100-200, the distribution becomes significantly tighter, indicating that we would get a good estimate of the population mean with samples of this size. For sample sizes of 500 and 1000, the sample means are very tightly clustered around the population mean, suggesting a highly reliable estimate. The plot indicates that increasing sample sizes beyond 300 provides diminishing returns in terms of reducing the variability of the sample mean, but still contributes to a more precise estimate.
